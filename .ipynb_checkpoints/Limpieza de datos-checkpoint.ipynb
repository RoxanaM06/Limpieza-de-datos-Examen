### Limpieza de las variables numéticas

#### - Limpieza de la columna "Customer_Age"

##### * Clasificamos los clientes que compran cada categoría de tarjeta de crédito según las edades porque así podemos ver entre cuales edades están comprando cada tipo de tarjeta de crédito.
##### * Sustituimos los valores atípicos por la media (según instrucciones dadas).

```python
edad = tarjeta_credito.loc[tarjeta_credito['Card_Category'] == 'Blue', ['Customer_Age']]
```

```python
plt.boxplot(edad)
```

```python
edad2 = tarjeta_credito.loc[tarjeta_credito['Card_Category'] == 'Silver', ['Customer_Age']]
```

```python
plt.boxplot(edad2)
```

```python
tarjeta_credito_q1 = tarjeta_credito.loc[:, ['Card_Category','Customer_Age']].groupby(['Card_Category'], as_index = False).quantile(0.25)
tarjeta_credito_q1.columns = ['Card_Category','Customer_Age_q1']
```

```python
tarjeta_credito_q3 = tarjeta_credito.loc[:, ['Card_Category','Customer_Age']].groupby(['Card_Category'], as_index = False).quantile(0.75)
tarjeta_credito_q3.columns = ['Card_Category','Customer_Age_q3']
```

```python
tarjeta_credito_q2 = tarjeta_credito.loc[:, ['Card_Category','Customer_Age']].groupby(['Card_Category'], as_index = False).mean()
tarjeta_credito_q2.columns = ['Card_Category','Customer_Age_q2']
```

```python
tarjeta_credito_iqr = tarjeta_credito_q1.merge(tarjeta_credito_q3, on = 'Card_Category', how = 'inner').merge(tarjeta_credito_q2, on = 'Card_Category', how = 'inner')
```

```python
tarjeta_credito_iqr
```

```python
tarjeta_credito_iqr['iqr'] = tarjeta_credito_iqr.apply(lambda r: r['Customer_Age_q3'] - r['Customer_Age_q1'], axis = 1)
```

```python
tarjeta_credito_iqr
```

```python
tarjeta_credito_iqr['upper'] = tarjeta_credito_iqr.apply(lambda r: (r['Customer_Age_q3'] + 1.5*r['iqr']), axis = 1)
tarjeta_credito_iqr['lower'] = tarjeta_credito_iqr.apply(lambda r: (r['Customer_Age_q1'] - 1.5*r['iqr']), axis = 1)
```

```python
tarjeta_credito_iqr
```

```python
tarjeta_credito_t = tarjeta_credito.merge(tarjeta_credito_iqr, on = 'Card_Category', how = 'left')
```

```python
tarjeta_credito_t.head(5)
```

```python
def getRealAge(r):
    if ( r["Customer_Age"] <= r["lower"] ) | ( r["Customer_Age"] >= r["upper"] ):
        return r["Customer_Age_q2"]
    else:
        return r["Customer_Age"]
```

```python
tarjeta_credito_t['New_Customer_Age'] = tarjeta_credito_t.apply( lambda r: getRealAge(r) , axis = 1 )
```

```python
tarjeta_credito_t.head(5)
```

##### A continuación podemos ver que si mostramos el boxplot de la edad de los clientes según la categoría Silver, ya no tenemos valores atípicos

```python
new_age = tarjeta_credito_t.loc[tarjeta_credito_t['Card_Category'] == 'Silver', ['New_Customer_Age']]
```

```python
plt.boxplot(new_age)
```

##### Eliminando columnas innecesarias

```python
tarjeta_credito_t = tarjeta_credito_t.drop(['Customer_Age','Customer_Age_q1','Customer_Age_q3','Customer_Age_q2','iqr','upper','lower'], axis = 1)
```

```python
tarjeta_credito_t.head(5)
```
